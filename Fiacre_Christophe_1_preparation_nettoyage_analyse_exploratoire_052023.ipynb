### Colonnes retenues en fonction de l'analyse PESTEL

Les colonnes que nous allons garder pour la construction de notre Dataset suivent la démarche de l'analyse PESTEL.
L'analyse PESTEL est un outil d'analyse stratégique qui permet d'identifier les facteurs macro-environnementaux qui peuvent influencer une entreprise ou une organisation. Les critères de l'analyse PESTEL sont les suivants :

**- Politique** : il s'agit de l'impact des politiques gouvernementales, des réglementations, des lois et des politiques fiscales sur l'entreprise. Cela peut inclure des facteurs tels que les politiques commerciales et les politiques de protection de l'environnement.

**- Économique** : il s'agit de l'impact des conditions économiques sur l'entreprise. Cela peut inclure des facteurs tels que le PIB, le taux de change, le niveau de croissance économique et le taux d'intérêt.

**- Social** : il s'agit de l'impact des changements sociaux sur l'entreprise. Cela peut inclure des facteurs tels que les attitudes et les comportements des consommateurs, les tendances démographiques, la diversité culturelle et les changements de style de vie.

**- Technologique** : il s'agit de l'impact des progrès technologiques sur l'entreprise. Cela peut inclure des facteurs tels que les avancées technologiques, les nouvelles inventions, les nouvelles méthodes de production, l'automatisation et l'impact d'Internet sur les entreprises.

**- Environnemental** : il s'agit de l'impact des facteurs environnementaux sur l'entreprise. Cela peut inclure des facteurs tels que le changement climatique, les catastrophes naturelles, la pollution, les préoccupations environnementales et les réglementations environnementales.

**- Légal** : il s'agit de l'impact des lois et réglementations sur l'entreprise. Cela peut inclure des facteurs tels que les lois sur la santé et la sécurité, les lois du travail, les lois sur la concurrence, les lois sur les droits d'auteur et les lois sur la protection des données.
